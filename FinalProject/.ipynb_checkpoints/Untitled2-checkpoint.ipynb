```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
# from statsmodel.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings

warnings.filterwarnings('ignore')
```

```python
temp = pd.read_csv('./city_temperature.csv')
```

```python
temp.head()
```

```python
temp.dtypes
```

```python
temp.shape
```

```python
temp = temp.drop(columns=["State", "Day", "Month","Country","City"])
temp
```

```python
temp.isnull().sum()
```

```python
temp['X-Mean'] = abs(temp['AvgTemperature'] - temp['AvgTemperature'].mean())
temp['1.96 * std'] = 1.96 * temp['AvgTemperature'].std()
temp['Outlier'] = temp['X-Mean'] > temp['1.96 * std']
```

```python
temp = temp.loc[temp['Outlier']==False]
```

```python
temp.shape
```

```python
temp.rename(columns = 
            {
             'AvgTemperature':'Avg_temp'
            }, inplace=True)
```

```python
temp.head()
```

```python
temp = temp.drop(columns=["X-Mean", "1.96 * std", "Outlier"])
temp
```

```python
temp['Date'] = pd.to_datetime(temp['Date'])
temp.set_index('Date', inplace=True)
temp.index
```

```python
temp.describe()
```

```python
latest_data = temp.loc['1900':'2019']
latest_data.head()
```

```python
latest_data[['Region','Avg_temp']].groupby(['Region']).mean().sort_values('Avg_temp')
```

```python
#ploting:
plt.figure(figsize=(9,4))
sns.lineplot(x='Year',y='Avg_temp', data=latest_data)
plt.show()
```

```python
resample_data = latest_data[['Avg_temp']].resample('A').mean()
```

```python
resample_data.head()
```

```python
resample_data.plot(title='Temperature changes from 1980 - 2019', figsize=(8,5))
plt.ylabel=('Temperature')
plt.xlabel=('Year')
plt.legend()
```

```python
from statsmodels.tsa.stattools import adfuller

print('Results of Dickey-Fuller Test:')

test_data = adfuller(resample_data['Avg_temp'], autolag='AIC')
data_output = pd.Series(test_data[:4], index=['Test Statistics','p-value','Lag Used','Number of Observation Used'])
for key, value in test_data[4].items():
    data_output['Critical Value (%s)' %key] = value
print(data_output)
```

```python
#Estimating trend

resample_data_logScale = np.log(resample_data)
plt.plot(resample_data_logScale)
```

```python
movingAverage = resample_data_logScale.rolling(window=3).mean()
movingSTD = resample_data_logScale.rolling(window=3).std()
plt.plot(resample_data_logScale)
plt.plot(movingAverage, color='red')
```

```python
resampleLogScaleMinusMovingAverage = resample_data_logScale - movingAverage
resampleLogScaleMinusMovingAverage.head(12)

#Remove NaN Values
resampleLogScaleMinusMovingAverage.dropna(inplace=True)
resampleLogScaleMinusMovingAverage.head()
```

```python
def test_stationarity(timeseries):
    
    #Determing Rolling Statistics
    movingAverage = timeseries.rolling(window=3,center=True).mean()
    movingSTD = timeseries.rolling(window=3,center=True).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
    std = plt.plot(movingSTD, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')

    test_data = adfuller(timeseries['Avg_temp'], autolag='AIC')
    data_output = pd.Series(test_data[:4], index=['Test Statistics','p-value','Lag Used','Number of Observation Used'])
    for key, value in test_data[4].items():
        data_output['Critical Value (%s)' %key] = value
    print(data_output)
#     fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,5))
#     ax1.plot(resample_data, label='Original')
#     ax1.plot(rol_mean, label='Rolling Mean')
#     ax1.plot(ewm, label='Exponentially weighted mean')
#     ax1.set_title('Temperature Change from 1980-2020')
#     ax1.set_ylabel('Temperature')
#     ax1.set_xlabel('Year')
#     ax1.legend()

#     ax2.plot(rol_std, label='Rolling STD')
#     ax2.set_title('Temperature Change from 1980-2020')
#     ax2.set_ylabel('Temperature')
#     ax2.set_xlabel('Year')
#     ax2.legend()

#     plt.tight_layout()
#     plt.show()
```

```python
test_stationarity(resampleLogScaleMinusMovingAverage)
```

```python
expDecayWeightAverage = resample_data_logScale.ewm(halflife=3,min_periods=0,adjust=True).mean()
plt.plot(resample_data_logScale)
plt.plot(expDecayWeightAverage, color='red')
```

```python
datasetLogScaleMinusMovingExponentialDecayAverage = resample_data_logScale - expDecayWeightAverage
test_stationarity(datasetLogScaleMinusMovingExponentialDecayAverage)
```

```python
datasetLogDiffShifting = resample_data_logScale - resample_data_logScale.shift()

plt.plot(datasetLogDiffShifting)
```

```python
datasetLogDiffShifting.dropna(inplace=True)
test_stationarity(datasetLogDiffShifting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(resample_data_logScale, freq=3)

trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid

plt.subplot(411)
plt.plot(resample_data)
plt.title('Original')
plt.figure(figsize=(6,5))

plt.subplot(412)
plt.plot(trend)
plt.title('Trend')
plt.figure(figsize=(6,5))

plt.subplot(413)
plt.plot(seasonal)
plt.title('Seasonal')
plt.figure(figsize=(6,5))

plt.subplot(414)
plt.plot(residual)
plt.title('Residual')
plt.figure(figsize=(6,5))

plt.tight_layout()
```

```python
decomposedLogData = residual
decomposedLogData.dropna(inplace=True)

#Determing Rolling Statistics
movingAverage = decomposedLogData.rolling(window=3,center=True).mean()
movingSTD = decomposedLogData.rolling(window=3,center=True).std()

#Plot rolling statistics:
orig = plt.plot(decomposedLogData, color='blue', label='Original')
mean = plt.plot(movingAverage, color='red', label='Rolling Mean')
std = plt.plot(movingSTD, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
    
print('Results of Dickey-Fuller Test:')

test_data = adfuller(decomposedLogData, autolag='AIC')
data_output = pd.Series(test_data[:4], index=['Test Statistics','p-value','Lag Used','Number of Observation Used'])
for key, value in test_data[4].items():
    data_output['Critical Value (%s)' %key] = value
print(data_output)
```

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(datasetLogDiffShifting, nlags=10)
lag_pacf = pacf(datasetLogDiffShifting, nlags=10, method='ols')

#plot acf:
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(datasetLogDiffShifting)),linestyle='--',color='gray')
plt.title('Partial correlation Function')
plt.tight_layout()
```

```python
from statsmodels.tsa.arima_model import ARIMA

#AR Model
model = ARIMA(resample_data_logScale, order=(2,1,2))
result_AR = model.fit(dis=-1)
plt.plot(datasetLogDiffShifting)
plt.plot(result_AR.fittedvalues, color='red')
plt.title('RSS: %4f' %sum((result_AR.fittedvalues-datasetLogDiffShifting['Avg_temp'])**2))
print('Plotting AR model')
```

```python
#MA Model
model = ARIMA(resample_data_logScale, order=(2,1,0))
result_AR = model.fit(dis=-1)
plt.plot(datasetLogDiffShifting)
plt.plot(result_AR.fittedvalues, color='red')
plt.title('RSS: %4f' %sum((result_AR.fittedvalues-datasetLogDiffShifting['Avg_temp'])**2))
print('Plotting AR model')
```
